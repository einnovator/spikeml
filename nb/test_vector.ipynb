# Upsample

```python



def test_upsample_zeros():
    x = np.array([1, 2, 3])
    y = upsample(x, R=3, method=UpsampleMethod.ZEROS)
    expected = np.array([1, 0, 0, 2, 0, 0, 3, 0, 0])
    assert np.array_equal(y, expected)


def test_upsample_repeat():
    x = np.array([1, 2, 3])
    y = upsample(x, R=3, method=UpsampleMethod.REPEAT)
    expected = np.array([1, 1, 1, 2, 2, 2, 3, 3, 3])
    assert np.array_equal(y, expected)


def test_upsample_normal_shape():
    x = np.array([1, 2, 3])
    y = upsample(x, R=4, method=UpsampleMethod.NORMAL, sigma=1.0)
    assert len(y) == len(x) * 4  # correct length


def test_upsample_normal_smoothness():
    # normal upsampled version should be smooth â€” no sudden jumps
    x = np.array([1.0, 2.0, 1.0])
    y = upsample(x, R=5, method=UpsampleMethod.NORMAL, sigma=1.0)
    diffs = np.diff(y)
    # The sign of differences should not change too frequently
    sign_changes = np.sum(np.diff(np.sign(diffs)) != 0)
    assert sign_changes < len(y) // 2  # rough smoothness heuristic


def test_upsample_invalid_R():
    with pytest.raises(ValueError):
        upsample(np.array([1, 2, 3]), R=0)


def test_upsample_invalid_method():
    with pytest.raises(ValueError):
        upsample(np.array([1, 2, 3]), R=2, method="invalid")


def test_upsample_all():
    test_upsample_zeros()
    test_upsample_repeat()
    test_upsample_normal_shape()
    test_upsample_normal_smoothness()
    test_upsample_invalid_R()
    test_upsample_invalid_method()
    
test_upsample_all()
```
